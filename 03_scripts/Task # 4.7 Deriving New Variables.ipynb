```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import the merged pickle dataset
orders_products_merged = pd.read_pickle(r'C:\Users\TemporaryAdmin\Desktop\yoana_learning\Data Analytics\Immension\tasks\my_tasks\part_4\02_data\prepared_data\orders_products_merged.pkl')
```

```python
# Creating a subset of the dataframe with a smaller number of rows
df = orders_products_merged[:1000000]
```

```python
df.head()
```

```python
# Dropping the unnamed columns in the beginning
df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0_x'])
```

```python
# Defining the price_label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying the price_label function 
df['price_range'] = df.apply(price_label, axis=1)
```


# Loops

```python
# Check how many orders are there on the different days of the week
orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Assign Saturday (=0) is the busiest day. Wednesday (=4) is the least busy day. Every other day is regularly busy. 
result = []

for value in orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check the new column with the assigned values
result
```

```python
# Assign a new column in the dataset = result column
orders_products_merged['busiest_day'] = result
```

```python
orders_products_merged['busiest_day'].value_counts(dropna = False)
```

# Task 4.7 Deriving New Variables

```python
# Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. 
result_updated = []

for value in orders_products_merged["orders_day_of_week"]:
  if value in (0, 1):
    result_updated.append("Busiest day")
  elif value in (4, 3) :
    result_updated.append("Least busy")
  else:
   result_updated.append("Regularly busy")
```

```python
result_updated
```

```python
orders_products_merged['busiest_day_updated'] = result_updated
```

```python
# Check the values of this new column for accuracy.
orders_products_merged['busiest_day_updated'].value_counts(dropna = False)
```

The number of the busiest days almost doubled now that we added Sunday to the Saturday. This is not surpring since they have close number of orders on both days (Sa with 6 252 941 and Su with 5 696 724). 
What is interesting is that the least busy days numbers changed drastically (We with 3 812 390) and by adding the next least busy day in the line - Tuesday we ended up with 7 681 137 orders which made the numbers for this category also almost doubled. 

```python
# The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”
# Check how many orders are there on the different times of the day
orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

The top 3 busiest hours are 9 am, 10 am and 1pm. 
The top 3 least busy hours are 1 am, 3 am and 2 am.

```python
# Create the if statement for the hours division
result_hours = []

for value in orders_products_merged["order_hour_of_day"]:
  if value in (10,11,14):
    result_hours.append("most_orders")
  elif value in (2,4,3) :
    result_hours.append("average_orders")
  else:
   result_hours.append("fewest_orders")
```

```python
result_hours
```

```python
# Create a new column with the results
orders_products_merged['busiest_period_of_day'] = result_hours
```

```python
# Check the values of this new column for accuracy.
orders_products_merged['busiest_day_updated'].value_counts(dropna = False)
```
