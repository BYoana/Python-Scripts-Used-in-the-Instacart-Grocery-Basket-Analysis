# Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

```python
ords_prods_merge['average_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spent'] < 10, 'spent_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spent'] >= 10, 'spent_flag'] = 'High Spender'
```

```python
ords_prods_merge['spent_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge.head(100)
```

Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 

```python
ords_prods_merge['regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['regularity'] > 20, 'regularity_shopping'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['regularity'] > 10 & (ords_prods_merge['regularity'] <= 20), 'regularity_shopping'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['regularity'] <= 10 & (ords_prods_merge['regularity'] <= 20), 'regularity_shopping'] = 'Regular customer'
```

```python
ords_prods_merge['regularity_shopping'].value_counts(dropna= False)
```
